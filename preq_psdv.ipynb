## 1.List Operations and Basic Statistics:
    Create a list of the first 20 natural numbers.
    Compute the sum of all even numbers in the list.
    Compute the product of all odd numbers in the list.
    Calculate the mean, median, and standard deviation of the list.

```python
import math

nums = list(range(1, 21))
print(f"List of first 20 natural numbers: {nums}")

even_sum = 0
for number in nums:
    if number%2 == 0:
        even_sum += number
print(f"Sum of all even numbers in the list: {even_sum}")

odd_product = math.prod(number for number in nums if number%2 != 0)
print(f"Product of all odd numbers in the list: {odd_product}")

mean = sum(nums)/ len(nums)
print(f"Mean: {mean}")

nums_sorted = sorted(nums)
n = len(nums)
if n % 2 == 1:
    median = nums_sorted[n // 2]
else:
    median = (nums_sorted[n // 2 - 1] + nums_sorted[n // 2]) /2
print(f"Median: {median}")

variance = sum((i - mean) ** 2 for i in nums) / (len(nums) - 1)
std_dev = variance ** 0.5
print(f"Standard Deviation: {std_dev}")
```

### Python data type is categorized into two types:

    Mutable Data Type – A mutable data type is one whose values can be changed.
    It is memory efficient as no new objects are created for frequent changes so it might be slow.
    Used when we need to modify, add or remove existing data frequently.
        Example: Lists, Dictionaries, and Sets

    Immutable Data Type – An immutable data type is one in which the values can’t be changed or altered.
    It might be faster than mutable data types as there is no need to track changes.
    Used when we want to ensure that the data remains consistent or unchanged.
        Example: Strings and Tuples

## 3. Fibonacci Sequence:

    Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
    Plot the Fibonacci sequence using a line graph.

```python
import matplotlib.pyplot as plt

def fibonacci_sequence(n):
    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence[:n]
#Example value for n
n = 20
result = fibonacci_sequence(n)
print(f"List of first {n} Fibonacci numbers: {result}")

def plot_fibonacci(n):
    fib_sequence = fibonacci_sequence(n)
    plt.plot(range(n), fib_sequence, marker='o', linestyle='-', color='green')
    plt.title(f'Fibonacci Sequence (First {n} Numbers)', loc = 'left')
    plt.xlabel('Number')
    plt.ylabel('Fibonacci Number')
    plt.grid(True)
    plt.show()
plot_fibonacci(20)
```

## 9. Merging Dictionaries:

    Write Python code to merge two dictionaries. If a key is present in both, sum their values.
    Example:
    dict1 = {'a': 10, 'b': 20}
    dict2 = {'b': 5, 'c': 15}
    Output: {'a': 10, 'b': 25, 'c': 15}

```python
def merge_dictionaries(d1, d2):
    merged_dict = d1.copy()
    for key, value in d2.items():
        if key in merged_dict:
            merged_dict[key] += value
        else:
            merged_dict[key] = value
    return merged_dict
#Example
d1 = {'a': 10, 'b': 20}
d2 = {'b': 5, 'c': 15}
result = merge_dictionaries(d1, d2)
print(result)
```

## 11. Remove Duplicates:

    Write a program to remove duplicate elements from a list while maintaining the order of elements.
    Calculate the mean and median of the unique elements.

```python
def remove_duplicates(lst):
    unique= list()
    for item in lst:
        if item not in unique:
            unique.append(item)
            
    mean = sum(unique) / len(unique)
    
    unique.sort()
    n = len(unique)
    if n % 2 == 1:
        median = unique[n // 2]
    else:
        median = (unique[n // 2 - 1] + unique[n // 2]) / 2

    return unique, mean, median

# Example
input_list = [1, 2, 2, 3, 3, 4, 5, 5, 5, 6, 7, 7]
unique, mean, median = remove_duplicates(input_list)

print("Unique Elements:", unique)
print("Mean:", mean)
print("Median:", median)
```

## 12. String Operations:

    Given a list of strings, write a Python function to:
    Find the longest string.
    Count how many strings start with a vowel.
    Visualize the distribution of string lengths using a histogram.

```python
import matplotlib.pyplot as plt

def str_operations(str_list):
    longest_string = max(str_list, key = len)

    vowels = ('a', 'e', 'i', 'o', 'u')
    count = sum(1 for s in str_list if s[0].lower() in vowels)

    length_of_str = [len(s) for s in str_list]

    plt.hist(str_list, bins = range(min(length_of_str), max(length_of_str) + 2, 1), color = 'green', edgecolor = 'black')
    plt.title("Distribution of strings")
    plt.xlabel("Length of Strings")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()

    return {
        'longest_string': longest_string,
        'count': count,
    }
    
# Example
strings = ["Firefox", "Safari", "Edge", "Chrome", "Opera", "Explorer", "UC Browser"]
results = str_operations(strings)

print(f"Longest String: {results['longest_string']}")
print(f"Number of Strings starting with a vowel: {results['count']}")
```

## 13. Sorting Dictionaries:

    Write a function to sort a list of dictionaries by a key.
    Example:
    data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
    Sort by age → [{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}]

```python
def sort_dict(data, sort_by_age):
    def sort_key(d):
        return d[sort_by_age]
        
    return sorted(data, key=sort_key)

# Example
data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
sorted_data = sort_dict(data, 'age')
print(sorted_data)
```

## 15. Anagram Finder:

    Write a Python function to find all anagrams of a word in a given list.
    Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]
    Output: ["enlist", "inlets"]

```python
def anagrams(word, words):
    sorted_words = ''.join(sorted(word))
    anagram = [word for word in words if ''.join(sorted(word)) == sorted_words]
    
    return anagram

# Example
word = "listen"
words = ["enlist", "google", "inlets", "banana"]
result = anagrams(word, words)

print(result)
```

## 18. Word Count and Frequency:

    Write a program to:
        Read a file and count the number of words in it.
        Find the most frequent word.
        Visualize the word frequencies using a bar chart.

```python
import matplotlib.pyplot as plt

filename = 'file.txt'
with open(filename, 'r') as file:
    text = file.read()
words = text.split()
num_words = len(words)
word_freq = {}

for word in words:
    w = word.lower()
    if w in word_freq:
        word_freq[w] += 1
    else:
        word_freq[w] = 1

most_freq_word = None
most_freq_count = 0

for word, count in word_freq.items():
    if count > most_freq_count:
        most_freq_word = word
        most_freq_count = count

top_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:10]
words, counts = zip(*top_words)

plt.bar(words, counts, color='green')
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 10 most frequent words")
plt.xticks(rotation=45)
plt.show()

print(f"Total number of words: {num_words}")
print(f"Most frequent word: '{most_freq_word}' with {most_freq_count} occurrences.")
```

## 19. Random Number Generation:

    Write a Python program to:
        Generate 100 random integers between 1 and 50.
        Save them in a text file, one per line.
        Read the file and compute the mean of the numbers.
        Visualize the distribution of the numbers using a histogram.

```python
import random
import matplotlib.pyplot as plt

def generate_random_numbers():
    numbers = [random.randint(1, 51) for _ in range(100)]
    return numbers

def save_numbers_to_file(numbers, filename):
    with open(filename, 'w') as file:
        for number in numbers:
            file.write(f"{number}\n")

def read_numbers_and_compute_mean(filename):
    with open(filename, 'r') as file:
        numbers = [int(line.strip()) for line in file]
    mean = sum(numbers) / len(numbers)
    return numbers, mean

def plot_histogram(numbers):
    plt.figure(figsize=(15, 9))
    plt.hist(numbers, bins=range(1, 51), edgecolor='black', color='green')
    plt.title('Distribution of Random Numbers')
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    plt.xticks(range(1, 51))
    
    plt.show()

def main():
    filename = 'random_numbers.txt'
    numbers = generate_random_numbers()
    save_numbers_to_file(numbers, filename)
    numbers_from_file, mean = read_numbers_and_compute_mean(filename)
    print(f"The mean of the numbers is: {mean}")
    plot_histogram(numbers_from_file)

main()
```

## 20. CSV File Operations:

    Write Python code to read a CSV file and compute the sum of values in a specific column.
    Visualize the data using a line plot.

```python
import csv
import matplotlib.pyplot as plt

def read_csv_file(filename):
    data = []
    with open(filename, mode='r') as file:
        csv_reader = csv.reader(file)
        headers = next(csv_reader)
        for row in csv_reader:
            data.append(row)
    return data, headers

def compute_column_sum(data, column_index):
    column_values = [float(row[column_index]) for row in data]
    return sum(column_values)

def plot_line_chart(data, column_index, headers):
    x_values = range(1, len(data) + 1)
    y_values = [float(row[column_index]) for row in data]
    plt.plot(x_values, y_values, marker='o', color='green')
    plt.title(f'Line Plot of {headers[column_index]}')
    plt.xlabel('Index')
    plt.ylabel(headers[column_index])
    plt.grid(True)
    plt.show()
    
def main():
    filename = 'day.csv'
    column_name = 'Numeric-2'
    data, headers = read_csv_file(filename)
    
    if column_name in headers:
        column_index = headers.index(column_name)
    else:
        print(f"Column '{column_name}' not found in CSV.")
        return
    
    column_sum = compute_column_sum(data, column_index)
    print(f"The sum of values in column '{column_name}' is: {column_sum}")
    plot_line_chart(data, column_index, headers)
    
main()
```

## 21. Basic Statistics:

    Write Python code to calculate:
        Mean, median, mode, and standard deviation of a list of numbers.
        Visualize the data using a box plot.

```python
import matplotlib.pyplot as plt

def calc_mean(numbers):
    return sum(numbers) / len(numbers)

def calc_median(numbers):
    numbers.sort()
    n = len(numbers)
    if n % 2 == 1:
        return numbers[n // 2]
    else:
        return (numbers[n // 2 - 1] + numbers[n // 2]) / 2

def calc_mode(numbers):
    frequency = {}
    for number in numbers:
        frequency[number] = frequency.get(number, 0) + 1
    max_freq = max(frequency.values())
    modes = [number for number, freq in frequency.items() if freq == max_freq] 
    if len(modes) == len(numbers):
        return "No unique mode"
    return modes

def calc_std_dev(numbers, mean):
    variance = sum((x - mean) ** 2 for x in numbers) / len(numbers)
    return variance ** 0.5

def plot_box_plot(numbers):
    plt.boxplot(numbers)
    plt.title('Box Plot of Numbers')
    plt.ylabel('Values')
    plt.show()

def main():
    numbers = list(range(1, 21))
    mean = calc_mean(numbers)
    median = calc_median(numbers)
    mode = calc_mode(numbers)
    std_dev = calc_std_dev(numbers, mean)

    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Mode: {mode}")
    print(f"Standard Deviation: {std_dev}")
    plot_box_plot(numbers)
    
main()
```

## 22. Normalization:

    Write a Python function to normalize a list of numbers to a range of 0 to 1.
    Visualize the normalized data using a line plot.

```python
import matplotlib.pyplot as plt

def normalize(numbers):
    min_value = min(numbers)
    max_value = max(numbers)
    return [(x - min_value) / (max_value - min_value) for x in numbers]

def plot_data(normalized_numbers):
    plt.plot(normalized_numbers, marker='o', color='green')
    plt.title("Normalized Data (Range 0 to 1)")
    plt.xlabel("Index")
    plt.ylabel("Normalized Value")
    plt.grid(True)
    plt.show()

def main():
    numbers = [12, 16, 20, 24, 28, 32, 36, 40, 44, 48, 52, 56, 60]
    normalized_numbers = normalize(numbers)
    print("Normalized Data:", normalized_numbers)
    plot_data(normalized_numbers)
main()
```

## 24. Grade Calculator:

    Write a Python program that takes a numerical grade (0-100) as input and prints the corresponding letter grade (A, B, C, D, F).

```python
grade = float(input("Enter a numerical grade (0-100): "))
def get_letter_grade(grade):
    if 90 <= grade <= 100:
        return "A"
    elif 80 <= grade < 90:
        return "B"
    elif 70 <= grade < 80:
        return "C"
    elif 60 <= grade < 70:
        return "D"
    elif 0 <= grade < 60:
        return "F"
    else:
        return "Invalid grade. Please enter a number between 0 and 100."
        
letter_grade = get_letter_grade(grade)
print(f"The letter grade is: {letter_grade}")
```

## 25. Leap Year Checker:

    Write a Python program that takes a year as input and checks if it is a leap year.

```python
year = int(input("Enter a year: "))
def leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        return True
    else:
        return False

if leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```
